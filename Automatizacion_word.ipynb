```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
doc = DocxTemplate("notas_alumnos.docx")
```

```python
nombre = "All GoHer"
telefono = "921921921"
correo = "tuprofe@gmail.com"
fecha = datetime.today().strftime("%d/%m/%Y")
```

```python
print(fecha)
```

```python
constantes = {"nombre":nombre, "telefono":telefono, "correo":correo, "fecha":fecha}
```

```python
print(constantes)
```

```python
df = pd.read_excel("notas.xlsx")
```

```python
df
```

```python
for indice, fila in df.iterrows():
    contenido = {"nombre_alumno":fila["Nombre del Alumno"], "nota_mate":fila["Matematica"],"nota_fisica":fila["Fisica"], "nota_quimica":fila["Quimica"]}
    contenido.update(constantes)
    doc.render(contenido)
    doc.save(f'Notas_de_{fila["Nombre del Alumno"]}.docx')
    print(contenido)
```
